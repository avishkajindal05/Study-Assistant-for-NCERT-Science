# Tokenizer Comparison
Comparing GPT-2 BPE, BERT WordPiece, and T5 SentencePiece on 5 representative passages from the corpus.

```python
pip install transformers
```

```python
from transformers import AutoTokenizer

tokenizers = {
    "GPT-2 BPE":        AutoTokenizer.from_pretrained("gpt2"),
    "BERT WordPiece":   AutoTokenizer.from_pretrained("bert-base-uncased"),
    "T5 SentencePiece": AutoTokenizer.from_pretrained("t5-small"),
}

passages = {
    "P1_definition":    "Displacement is the net change in the position of an object between the two given instants of time.",
    "P2_acceleration":  "The rate of change of velocity is called acceleration. If the velocity of an object changes from u to v in time t, then acceleration a = (v - u) / t.",
    "P3_circular":      "When an object moves in a circular path with constant speed its motion is called uniform circular motion.",
    "P4_equation":      "For the motion of an object in a straight line with constant acceleration v = u + at and s = ut + half at squared.",
    "P5_hyphenated":    "The velocity-time graph shows specific-heat-capacity and position-time relationship.",
}
```

```python
print(f"{'Passage':<20} {'GPT-2':>8} {'BERT':>8} {'T5':>8}")
print("-" * 48)
for name, text in passages.items():
    counts = {k: len(v.tokenize(text)) for k, v in tokenizers.items()}
    print(f"{name:<20} {counts['GPT-2 BPE']:>8} {counts['BERT WordPiece']:>8} {counts['T5 SentencePiece']:>8}")
```

```python
terms = ["velocity-time", "specific-heat-capacity", "position-time", "acceleration"]

print("How each tokenizer splits scientific terms:\n")
for term in terms:
    print(f"Term: '{term}'")
    for name, tok in tokenizers.items():
        tokens = tok.tokenize(term)
        print(f"  {name:<22}: {tokens}")
    print()
```

## Observations

**Token counts:** T5 consistently produces the most tokens (up to 45 for P2 vs
GPT-2's 37) because its Unigram LM model fragments scientific terms more aggressively.
GPT-2 and BERT are close on most passages.

**Key boundary disagreements:**

1. `acceleration` — BERT and T5 keep it as one token, GPT-2 splits it into
   `['ac', 'celer', 'ation']`. This means a BM25 index built with GPT-2 tokenization
   would treat a query "what is acceleration" differently from the corpus — the
   query token `acceleration` wouldn't match the corpus tokens `ac`, `celer`, `ation`.
   Silent retrieval failure.

2. `specific-heat-capacity` — T5 fragments it into 8 tokens including `'he'`, `'at'`,
   `'cap'`, `'a'`, `'city'`. A query containing this term would score near-zero
   against corpus chunks indexed with T5 tokenization.

3. T5's `▁` prefix on every first token (e.g. `▁velocity`) is a SentencePiece
   artifact representing a leading space. If you strip whitespace inconsistently
   between index and query time, `▁velocity` and `velocity` won't match.

**Why we use TextPreprocessor (simple whitespace + lowercase) for BM25:**
All three tokenizers introduce term-splitting behavior that would break tokenization
symmetry between corpus and query. Our custom preprocessor keeps terms intact and
is identical on both sides — the most important property for reliable BM25 scoring.
